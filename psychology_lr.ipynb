```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import sklearn
```

```python
df = pd.read_csv('_dane_psychologiczne.csv')
df
```

```python
del df['ID']
```

```python
graph = sns.scatterplot( data = df, x = 'Godziny_snu', y = 'Poziom_stresu', color='#A2007B')
# linear dependence
```

```python
X1 = df['Godziny_snu']
y = df['Poziom_stresu']
m = len(X1)
print(X1)
print(y)
m
```

```python
#zmieniamy rozmiar
X1 = X1.values.reshape((m,1))
print(X1)
print(X1.shape)
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X1,y)
```

```python
print(regression.intercept_)
print(regression.coef_[0])
```
