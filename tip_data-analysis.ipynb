**Tips data analysis**

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
a = sns.relplot(x='total_bill', y='tip', data=df) # w ten sposb tworzymy wykres punktowy
a
```

```python
sns.set(font_scale=1.5) #zwiksza/zmniejsza napisy
a = sns.relplot(x='total_bill', y='tip', data=df) # w ten sposb tworzymy wykres punktowy
a
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size') #wprowadzamy 3 zmienn
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time') #wprowadzamy 3 zmienn
#col dzieli wykresy na x
# row zamaist col da wykresy w pionie
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time', row='smoker')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', hue='day') # hue - kolory w zalenoci od ...
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', hue='day', kind='swarm')
# kind = 'swarm' rozszerza
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', hue='day', kind='box')
# kind = 'box' wykresy pudekowe
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', hue='day', kind='violin')
# kind = 'violin' wykresy wieloktne
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', hue='day', kind='bar')
# kind = 'bar' wykres supkowy
```

```python
_ = sns.catplot(data=df, x='day', hue='day', kind='count')
# kind = 'count' licznik
```
